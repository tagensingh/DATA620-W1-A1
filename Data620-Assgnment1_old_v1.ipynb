```python
# Importing the required libraries
import sys
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()
```

```python
print(' To demonstrate that this install is functional - See the Python Version info below')
sys.version
```

```python
G.clear()

# Adding the nodes per the assignment instructions

####
# TNS - Comment - I commented after learning that by creating the edges the vertices are automatically created.
#G.add_nodes_from (["Andre","Beverly","Carol","Diane","Ed","Fernando","Garth","Heather","Ike","Jane"])
####

# Adding the edges and nodes per the assignment instructions

G.add_edge("Jane","Ike")
G.add_edge("Ike","Heather")

G.add_edge("Heather","Fernando")
G.add_edge("Heather","Garth")

G.add_edge("Fernando","Garth")
G.add_edge("Fernando","Carol")
G.add_edge("Fernando","Andre")
G.add_edge("Fernando","Diane")

G.add_edge("Garth","Diane")
G.add_edge("Garth","Beverly")
G.add_edge("Garth","Ed")

G.add_edge("Diane","Beverly")
G.add_edge("Diane","Andre")
G.add_edge("Diane", "Ed")
G.add_edge("Diane", "Carol")

G.add_edge("Ed","Beverly")
G.add_edge("Carol","Andre")
G.add_edge("Andre", "Beverly")
```

```python
# Setting up the plot.

nx.draw(G, with_labels = True,node_shape="s",linewidths=5,node_color="r")
plt.title(' A Simple Krackhardt Kite of 10 Nodes')
plt.show()
```
